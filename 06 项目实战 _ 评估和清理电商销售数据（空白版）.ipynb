# 项目：评估和清理英国电商公司销售数据

## 分析目标

此数据分析的目的是，根据市场销售数据，挖掘畅销产品，以便制定更有效的市场策略来提升营收。

本实战项目的目的在于练习评估数据干净和整洁度，并且基于评估结果，对数据进行清洗，从而得到可供下一步分析的数据。

## 简介

原始数据集记录了一家英国在线零售公司在2010年12月1日至2011年12月9日期间的所有交易情况，涵盖了该公司在全球不同国家和地区的业务数据。该公司主要销售覆盖各个场景的礼品，包括但不限于生日礼品、结婚纪念品、圣诞礼品等等。该公司的客户群体主要包括批发商和个人消费者，其中批发商占据了相当大的比例。

数据每列的含义如下：
- `InvoiceNo`: 发票号码。6位数，作为交易的唯一标识符。如果这个代码以字母“c”开头，表示这笔交易被取消。
- `StockCode`: 产品代码。5位数，作为产品的唯一标识符。
- `Description`: 产品名称。
- `Quantity`: 产品在交易中的数量。
- `InvoiceDate`: 发票日期和时间。交易发生的日期和时间。
- `UnitPrice`: 单价。价格单位为英镑（£）。
- `CustomerID`: 客户编号。5位数，作为客户的唯一标识符。
- `Country`: 国家名称。客户所居住的国家的名称。

## 读取数据

```python
# 加载包
import pandas as pd

# 加载数据
original_data = pd.read_csv('E:/file/data_analysis/python/小白玩转python/随课数据/数据评估和数据清洗篇/e_commerce.csv')
original_data.head(50)
```

## 评估数据

### 评估整洁度

```python

original_data.head(10) # 前10
original_data.sample(10) # 随机抽取10个数
```

从抽样的结果来看，数据符合“每列是一个变量，每行是一个观察值，每个单元格是一个值”，具体来看，每行代表一次商品交易，每列是交易的相关变量，因此不存在结构性问题

### 评估整洁干净度

```python
original_data.info()
```

从输出结果来看，数据共有541909条观察值，其中`Description`和`CustomerID`变量有缺失值，InvoiceDate的数据类型为应为日期，CustomerID列的数据变量类型应为字符串，这两列需进行数据转换

### 评估缺失数据

```python
# 找出Description列所有缺失的行
original_data[original_data['Description'].isnull()]
```

从输出结果来看，这些缺失Description的行，对应的UnitPrice行全为0，为验证这些猜想，增加筛选条件，看是否存在Description行缺失时UnitPrice却不为零的情况

```python
original_data[(original_data['Description'].isnull()) & (original_data['UnitPrice'] != 0)]
```

从上面的输出结果来看，说明这些缺失Description的行，对应的UnitPrice行全为0。   
Description表示商品名称，UnitPrice表示商品价格，都是后续进行商品交易分析的重要变量，如果他们同时缺失，我们认为数据无法提供有效含义，因此这些数据后续可以被删除

`CustomerID`也存在缺失值，因此先根据条件提取出观察值

```python
original_data[original_data['CustomerID'].isnull()]
```

`CustomerID`表示客户编号，不是分析畅销产品的必要变量，而且从输出结果来看，虽然`CustomerID`为空，其他的数据仍有效，因此保留此变量为空的值

### 评估重复数据

根据变量的含义来看，虽然`InvoiceNo`、`StockCode`、`CustomerID`都是唯一标识符，但一次交易可能存在多个商品，所以`InvoiceNo`可以重复；不同交易可能包含同件商品，因此`StockCode`可以重复；顾客可以多次交易购买商品，因此`CustomerID`也可以重复。

### 评估不一致数据

不一致数据可能存在于`Country`变量中，我们要查看是否存在多个不同的值指代同一国家的情况

```python
original_data['Country'].value_counts()
```

从`Country`变量值来看，'USA'、'United States '均表示美国，'United Kingdom'、'UK'、'U.K.'均表示英国，因此应该在数据清洗阶段对这些值进行统一，只保留一个代称。

### 评估无效或错误数据

可以通过`describe()`方法，对数值统计信息进行快速了解

```python
original_data.describe()
```

从输出结果来看，`Quantity`和 `UnitPrice`存在负数，会对后续的分析产生影响，因此先筛选出`Quantity`数值为负数的观察值，进一步评估其含义

```python
original_data[original_data['Quantity']<0]
```

从删选条件来看，`Quantity`变量为负的观察值，`InvoiceNo`似乎均以C开头，表示订单被取消，后续进行验证此猜想

```python
original_data[(original_data['Quantity']<0) & (original_data['InvoiceNo'].str[0] != 'C')]
```

上述猜想被证明错误，但从结果来看，这些删选出来观察值`UnitPrice`均为0，因此增加`UnitPrice`的条件进行验证

```python
original_data[(original_data['Quantity']<0) & (original_data['InvoiceNo'].str[0] != 'C') & (original_data['UnitPrice'] != 0)]
```

根据输出结果，猜想得到验证，当`Quantity`变量为负数的时候，观察值满足以下条件之一：
- `InvoiceNo`以C开头，表示订单被取消
- `UnitPrice`为0，说明单价为0英镑

这些交易数据均不是有效成交数据，不贡献销售，不在后续分析范围内，因此我们将在数据清理步骤，将`Quantity`为负数的观察值删除。

接下来筛选出`UnitPrice`为负数的行

```python
original_data[original_data['UnitPrice'] < 0]
```

从输出结果来看，`UnitPrice`为负数的观察值都是坏账调整，不属于实际的商品交易数据，因此也在数据清理的步骤中将其删除

## 清洗数据

根据前面评估的部分可得到以下结论，我们需进行的清理步骤包括
- 把`InvoiceDate`变量的数据类型转换为为日期时间
- 把`CustomerID`变量的数据类型转换为字符串
- 把`Description`变量缺失的观察值删除
- 把`Country`变量值`"USA"`替换为`"United States"`
- 把`Country`变量值`"UK"`、`"U.K."`替换为`"United Kingdom"`
- 把`Quantity`变量值为负数的观察值删除
- 把`UnitPrice`变量值为负数的观察值删除

为区分清理前后的数据，将清理后的数据命名为`cleaned_data`

```python
# 拷贝原始变量

cleaned_date  = original_data.copy()
cleaned_date.head()
```

把`InvoiceDate`变量的数据类型转换为为日期时间

```python
# 利用pandas中的to_datetime()函数进行时间格式调整
cleaned_date['InvoiceDate']  = pd.to_datetime(cleaned_date['InvoiceDate'])
cleaned_date['InvoiceDate'] 
```

把`CustomerID`变量的数据类型转换为字符串

```python
cleaned_date['CustomerID'] = cleaned_date['CustomerID'].astype(str)
cleaned_date['CustomerID']
```

把`CustomerID`变量结尾的'.0'删除

```python
cleaned_date['CustomerID'] = cleaned_date['CustomerID'].str.slice(0,-2) 
cleaned_date['CustomerID']
```

把`Description`变量缺失的观察值删除

```python
# cleaned_date = cleaned_date.dropna(subset=['Description'])
cleaned_date.dropna(subset=['Description'],inplace=True)
```

```python
# 检查是否全部修改成功
len(cleaned_date[cleaned_date['Country'] == 'USA'])
```

```python
cleaned_date['Country'].replace(['UK',"U.K."],"United Kingdom",inplace=True)
```

```python
# 检查是否全部修改成功
print(len(cleaned_date[cleaned_date['Country'] == 'UK']))
print(len(cleaned_date[cleaned_date['Country'] == 'U.K.']))
```

把`Quantity`变量值为负数的观察值删除  

```python
cleaned_date = cleaned_date[cleaned_date['Quantity'] >= 0]
```

```python
# 检查是否还有负值
len(cleaned_date[cleaned_date['Quantity']<0])
```
